## Epidemic Learning

L’apprentissage épidémique (Epidemic Learning) est une variante de l’apprentissage
distribué où les nœuds d’un réseau échangent leurs modèles de manière aléa-
toire en suivant un protocole inspiré de la diffusion des épidémies. Contraire-
ment aux méthodes centralisées, cette approche ne repose pas sur un serveur
central, ce qui réduit la congestion et améliore la scalabilité.


### Configuration de l’Expérience

- Dataset: CIFAR-10
- Modèle: LeNet
- Optimiseur: SGD avec un taux d’apprentissage de 0.001
- Loss Function: CrossEntropyLoss
- Taille du Batch: 10 
- Nombre d’itérations: 200
- Configuration du Réseau : Fully Connected
